### Project goal

The goal is to create a model which automatically finds digits from MNIST dataset that are ambiguous (or in other
words, hard to classify).



```python
import utils

train_X, train_y, test_X, test_y = utils.load_mnist()
```

```python
input_shape = (train_X.shape[1],train_X.shape[2],train_X.shape[3])
test_eval = utils.batch_run_models(n_runs=30, 
                                   train_X=train_X, 
                                   train_y=train_y,
                                   test_X=test_X,
                                   test_y=test_y,
                                   input_shape=input_shape, 
                                   n_classes=10, 
                                   batch_size=128, 
                                   epochs=3)
```

```python
import pickle, utils

# Load previously saved model evaluation data
with open("test_eval_raw.pkl", 'rb') as f:
    test_eval = pickle.load(f)

utils.save_ambiguous_images(test_eval, image_folder_path="images")
```
